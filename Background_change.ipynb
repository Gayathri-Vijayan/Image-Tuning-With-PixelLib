```python
import pixellib
import os
from pixellib.tune_bg import alter_bg
from matplotlib import image
from matplotlib import pyplot
```

```python
change_bg = alter_bg(model_type = "pb")
change_bg.load_pascalvoc_model("C:\\Users\\gayat\\Downloads\\xception_pascalvoc.pb")
```

```python
change_bg.color_bg("view1.jpeg", colors = (0, 0, 0), output_image_name="1.jpg", detect = "car")
change_bg.color_bg("view2.jpeg", colors = (0, 0, 0), output_image_name="2.jpg", detect = "car")
change_bg.color_bg("view3.jpeg", colors = (0, 0, 0), output_image_name="3.jpg", detect = "car")
```

```python
image1= image.imread("1.jpg")
image2= image.imread("2.jpg")
image3= image.imread("3.jpg")
pyplot.imshow(image1)
pyplot.show()
pyplot.imshow(image2)
pyplot.show()
pyplot.imshow(image3)
pyplot.show()
```

```python
change_bg.change_bg_img(f_image_path = "view1.jpeg",b_image_path = "back.jpg", output_image_name="new_img1.jpg")
change_bg.change_bg_img(f_image_path = "view2.jpeg",b_image_path = "back.jpg", output_image_name="new_img2.jpg")
change_bg.change_bg_img(f_image_path = "view3.jpeg",b_image_path = "back.jpg", output_image_name="new_img3.jpg")
```

```python
image4 = image.imread("new_img1.jpg")
image5 = image.imread("new_img2.jpg")
image6 = image.imread("new_img3.jpg")
pyplot.imshow(image4)
pyplot.show()
pyplot.imshow(image5)
pyplot.show()
pyplot.imshow(image6)
pyplot.show()
```
